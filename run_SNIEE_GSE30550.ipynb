```python
import pandas as pd
df = pd.read_csv('data/GSE30550/landscape.csv', index_col=0)
local_net = df.index[:13].to_list()
sns.heatmap(df)
```

```python
gene2i = {x: i for i, x in enumerate(sniee.adata.var_names)}

res = []
print(local_net)
for p in local_net:
    gene1, gene2 = p.split('-')
    gene1_i, gene2_i = gene2i[gene1], gene2i[gene2]

    for key in sniee.entropy_dict.keys():
        val = sniee.entropy_dict[key][gene1_i, gene2_i]
        res.append([p, gene1_i, gene2_i, key, val])

df = pd.DataFrame(res, columns=['relation', 'gene1_i', 'gene2_i', 'sample', 'score'])
df = pd.pivot(df, index='sample', columns='relation', values='score')
sns.heatmap(df.T)
```

```python
adata = sniee.adata.copy()
adata = adata[df.index]
for col in df.columns:
    adata.obs[col] = df[col].to_dict()
adata.obs.columns
```

```python
adata.obs['local_score'] = adata.obs[local_net].sum(axis=1)

sns.lineplot(adata.obs, x='time', y='local_score', units='person', hue='clinic_pheno:ch1', estimator=None)
```

```python
import matplotlib.pyplot as plt

for person in adata.obs['person'].unique():
    sadata = adata[adata.obs['person'] == person]
    df = sadata.obs[local_net]
    df.index = sadata.obs.time
    sns.heatmap(df.T)
    plt.title(f'landscape plot for subject {person}')
    plt.show()
```
